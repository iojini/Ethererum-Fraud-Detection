```python
#Import pandas, matplotlib.pyplot, and seaborn 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the ethereum transaction data
df = pd.read_csv('transaction_dataset.csv')
print(df.shape)
```

```python
#Print the first several rows of the data
df.head()
```

```python
#Check columns
df.columns
```

```python
#Remove space between apostrophe and 1st letter for last 25 columns
df.columns = [x.strip() for x in df.columns]
df.columns
```

```python
#Remove unnamed:0 and index columns
df = df.drop(columns=["Unnamed: 0", "Index"])
df.head()
```

```python
#Determine the columns that have missing values
missing = pd.concat([df.isnull().sum(), 100 * df.isnull().mean()], axis=1)
missing.columns=['count', '%']
missing.sort_values(by='count')
```

25 out of 47 data columns have missing values. The last column, ERC20_most_rec_token_type, has the most missing values (i.e., 8.7%). Fortunately, there are no missing values in the target column. Since missing values make up less than 10% of the data, these will be dropped.

```python
#Drop rows with missing values 
df = df.dropna()
df
```

```python
#Examine dataset for duplicate values
duplicateRows = df[df.duplicated()]
print(duplicateRows)
```

There appears to be 18 duplicate addresses. The duplicated addresses identified will need to be removed.

```python
#Drop duplicates
df = df.drop_duplicates()
df
```

The dataset now has 8963 rows. 

```python
#Examine summary statistics for numeric columns
df.describe().T
```

The following 7 columns appears to consist of all 0s and will have to be deleted: (1) ERC20 avg time between sent tnx, (2) ERC20 avg time between rec tnx, (3) ERC20 avg time between rec 2 tnx, (4) ERC20 avg time between contract tnx, (5) ERC20 min val sent contract, (6) ERC20 max val sent contract, and (7) ERC20 avg val sent contract.

```python
#Remove selected columns
df = df.drop(columns=["ERC20 avg time between sent tnx", "ERC20 avg time between rec tnx", "ERC20 avg time between rec 2 tnx", "ERC20 avg time between contract tnx", "ERC20 min val sent contract", "ERC20 max val sent contract", "ERC20 avg val sent contract"])
df.head()
```

```python
#Shape of dataset 
df.shape
```

The dataset is now ready for next steps (i.e., EDA).

```python
#save the data to a new csv file
df.to_csv('transaction_dataset_cleaned.csv', index= False)
```
